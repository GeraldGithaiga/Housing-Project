<h4> Remove Outliers </h4>

```python
low, high = df["price"].quantile([0.1, 0.9])
mask_price = df["price"].between(low, high)

df = df[mask_price]

low, high = df["area"].quantile([0.1, 0.9])
mask_area = df["area"].between(low, high)

df = df[mask_area]
```

```python
#Percentage of retained data
total_records= df.shape[0]

retained_data = round(100*(len(df)/total_records),2)
print(retained_data,"%")
```

```python
import warnings
from glob import glob


from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.pipeline import make_pipeline
from sklearn.utils.validation import check_is_fitted
```

```python
# from category_encoders import OneHotEncoder
```

<h3> Split Data </h3>

```python
X= df.drop(columns=['price'])
y = df['price']
```

```python
##split into train and test

from sklearn.model_selection import train_test_split
np.random.seed(0)
#df_train, df_test = train_test_split(df, train_size=0.7, test_size = 0.3, random_state=100)

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.1, random_state=42)
```

```python

model = LinearRegression().fit(X_train,y_train)
print(model.__dict__)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

<h3> Baseline </h3>

```python
#Calculate the baseline mean absolute error for your model
y_mean = y_train.mean()
y_pred_baseline = [y_mean]*len(y_train)
baseline_mae = mean_absolute_error(y_train,y_pred_baseline)
print("Mean apt price:", y_mean)
print("Baseline MAE:", baseline_mae)
```

<h3>Iterate </h3>

```python
# Build Model
model = make_pipeline(

    SimpleImputer()
    ,Ridge())
# Fit model
model.fit(X_train,y_train)
```

```python
# retrieve intercept
intercept = model.named_steps["ridge"].intercept_

# retrieve coefficients
coefficients = model.named_steps["ridge"].coef_

# retrieve names
rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)
print(rf_model.__dict__)
```

```python
feature_importances = rf_model.feature_importances_
# create a series of names and values
for i, feature in enumerate(X):
    print(f"{feature}: {feature_importances[i]}")
```

```python
import pandas as pd
y_pred = model.predict(X_test)
y_pred_rounded = pd.Series(y_pred).round(2)
df = pd.DataFrame({"Actual": y_test, "Predicted": y_pred})
df.head()
```

```python
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', label='Line of Best Fit')

plt.xlabel('Actual Home Price Index')
plt.ylabel('Predicted Home Price Index')
plt.title('Actual vs Predicted Home Price Index')
plt.legend()
plt.grid(True)
plt.show()
```
